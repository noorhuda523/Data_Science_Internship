```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('customer_spending_score.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Date_of_Purchase']= pd.date_range(start='2024-01-01', end='2024-12-01', periods=len(df))
df.head()
```

```python
df['Customer_ID']=range(1, len(df)+1)
df.head()
```

```python
df['Annual Income (k$)']=pd.to_numeric(df['Annual Income (k$)'], errors='coerce')
df['Age']=pd.to_numeric(df['Age'], errors='coerce')
df['Spending Score (1-100)']=pd.to_numeric(df['Spending Score (1-100)'], errors='coerce')
df['Date_of_Purchase']=pd.to_datetime(df['Date_of_Purchase'], errors='coerce').dt.date
df
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates(keep='first')
```

```python
df=df.rename(columns={
    'Annual Income (k$)':'Annual_Income',
     'Spending Score (1-100)':'Spending_Score'
})
df
```

```python
df.columns.tolist()
```

```python
customer=df[(df['Spending_Score']>70) & (df['Age']<30)]
customer.sort_values(by='Annual_Income', ascending=False)
```
